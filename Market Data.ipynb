```python
data = %sql SELECT competitor, SUM(competitor_revenue) AS total_revenue FROM market_data GROUP BY competitor ORDER BY total_revenue DESC;
df = data.DataFrame()

# Create the barplot without hue
plt.figure(figsize=(10, 6))
g = sns.barplot(data=df, y='competitor', x='total_revenue', palette="deep",
                hue='competitor', legend=False)

# Set axis labels and title
g.set_xlabel("Revenue")
g.set_ylabel("Competitor")
plt.title("Total Competitor Revenue".upper())

# Show plot
plt.show()
```

```python
data = %sql WITH competitorCTE AS (SELECT product, competitor, MAX(competitor_revenue) as max_revenue, ROW_NUMBER () OVER(PARTITION BY product ORDER BY MAX(competitor_revenue) DESC) as row_num FROM market_data GROUP BY product, competitor) SELECT product, competitor, max_revenue FROM `competitorCTE`;
df = data.DataFrame()

# Initialize the Seaborn theme
sns.set_theme(style="whitegrid")

# Draw a nested barplot by product and competitor
g = sns.catplot(
    data=df, kind="bar",
    x="product", y="max_revenue", hue="competitor", 
    palette="pastel", height=6
)

# Customize the plot
g.despine(left=True)
g.set_axis_labels("Product", "Revenue")
g.legend.set_title("Competitor")
plt.title("Maximum Revenue Generated by Competitor and Product".upper())


# Show the plot
plt.show()
```

```python
data = %sql WITH productCTE AS(SELECT product, competitor, SUM(competitor_revenue) as total_revenue, ROW_NUMBER () OVER(PARTITION BY product ORDER BY SUM(competitor_revenue) DESC) as row_num FROM market_data GROUP BY product, competitor) SELECT product,competitor,total_revenue FROM `productCTE`;
df = data.DataFrame()

# Initialize the Seaborn theme
sns.set_theme(style="whitegrid")

# Draw a nested barplot by product and competitor
g = sns.catplot(
    data=df, kind="bar",
    x="product", y="total_revenue", hue="competitor", 
    palette="viridis", height=6
)

# Customize the plot
g.despine(left=True)
g.set_axis_labels("Product", "Revenue")
g.legend.set_title("Competitor")
plt.title("Total Revenue by Product and Competitor".upper())


# Show the plot
plt.show()
```

```python
data = %sql SELECT product, SUM(competitor_revenue) as revenue FROM market_data GROUP BY product ORDER BY revenue DESC;
df = data.DataFrame()

# Create the barplot without hue
plt.figure(figsize=(10, 6))
g = sns.barplot(data=df, y='product', x='revenue', palette="deep",
                hue='product', legend=False)

# Set axis labels and title
g.set_xlabel("Revenue")
g.set_ylabel("Product")
plt.title("Total Product Revenue by Competiitor".upper())

# Show plot
plt.show()
```

```python
data = %sql WITH monthsCTE AS(SELECT MONTHNAME(date) as months, product, SUM(competitor_revenue) as total_revenue, MONTH(date) as month_num, ROW_NUMBER() OVER(PARTITION BY product ORDER BY MONTH(date)) as row_num FROM market_data WHERE YEAR(`date`) = 2023 GROUP BY MONTHNAME(date),product,MONTH(date)) SELECT months, product, total_revenue FROM `monthsCTE`;
df = data.DataFrame()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))
g = sns.lineplot(data=df, x="months", y="total_revenue", hue="product",palette="deep")

# Set axis labels and title
g.set_xlabel("Months")
g.set_ylabel("Revenue")
g.legend(title="Product")
plt.title("Total Competitor Revenue by Product for 2023".upper())
```

```python
data = %sql WITH monthsCTE AS(SELECT MONTHNAME(date) as months, product, SUM(competitor_revenue) as total_revenue, MONTH(date) as month_num, ROW_NUMBER() OVER(PARTITION BY product ORDER BY MONTH(date)) as row_num FROM market_data WHERE YEAR(`date`) = 2024 GROUP BY MONTHNAME(date),product,MONTH(date)) SELECT months, product, total_revenue FROM `monthsCTE`;
df = data.DataFrame()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))
g = sns.lineplot(data=df, x="months", y="total_revenue", hue="product",palette="deep")

# Set axis labels and title
g.set_xlabel("Months")
g.set_ylabel("Revenue")
g.legend(title="Product")
plt.title("Total Competitor Revenue by Product for 2024".upper())
```

```python
data = %sql SELECT YEAR(date) as years, competitor, SUM(competitor_revenue) as total_revenue, ROW_NUMBER() OVER(PARTITION BY YEAR(date) ORDER BY SUM(competitor_revenue) DESC) as row_num FROM market_data GROUP BY YEAR(date), competitor;
df = data.DataFrame()

# Initialize the Seaborn theme
sns.set_theme(style="whitegrid")

# Draw a nested barplot by product and competitor
plt.figure(figsize=(8, 8))
g = sns.catplot(
    data=df, kind="bar",
    x="years", y="total_revenue", hue="competitor", 
    palette="colorblind", height=8
)

# Customize the plot
g.despine(left=True)
g.set_axis_labels("Years", "Revenue")
g.legend.set_title("Competitor")
plt.title("Revenue Distribution by Competitor and Year".upper())


# Show the plot
plt.show()
```

```python
# Run the SQL query and load data into a DataFrame
data = %sql SELECT competitor, SUM(competitor_revenue) AS total_revenue, SUM(competitor_revenue) * 100.0 / (SELECT SUM(competitor_revenue) FROM market_data) AS revenue_percentage FROM market_data GROUP BY competitor;
df = data.DataFrame()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['revenue_percentage'], labels=df['competitor'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(df)))

# Title and display
plt.title("Competitor Market Share Percentage Distribution".upper())
plt.show()
```

```python
data = %sql SELECT competitor, ROUND(AVG(customer_satisfaction),2) as avg_satisfaction_value FROM market_data GROUP BY competitor;
df = data.DataFrame()

# Create the barplot without hue
plt.figure(figsize=(12, 6))
g = sns.barplot(data=df, x='competitor', y='avg_satisfaction_value', palette="pastel",
                hue='competitor', legend=False)

# Set axis labels and title
g.set_xlabel("Competitor")
g.set_ylabel("Average Customer Satisfaction Value")
plt.title("Customer Satisfaction Value by Competitor".upper())

# Show plot
plt.show()
```
